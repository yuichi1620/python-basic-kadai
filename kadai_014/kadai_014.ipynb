```python
# 商品の価格を変数に代入する    
price1 =   100
price2 =   200

# 消費税10%を加えた合計金額を計算して返す関数
def total():
    tax = 1.1
    return (price1 + price2) * tax

# 関数を呼び出して合計金額を出力する
print(total())
```

```python
price1 = 100
price2 = 200

def total(tax):
    return (price1 + price2) * tax

print(total(1.1))
```

```python
price1 = 100
price2 = 200
tax = 1.1

def total():
    return price1 + price2

print(total() * tax)
```
