```python
class Human:
    def __init__(self, name: str, age: int):
        # 属性を定義
        self.name = name 
        self.age = age
        

    def printinfo(self):
        # neme属性とage属性を出力するメソッド
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")

# Humanクラスのインスタンスを生成
human = Human("雄一", 29)

# printinfoメソッドを呼び出す
human.printinfo()
```
